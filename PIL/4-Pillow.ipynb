```python
#load and show an image with Pillow
from PIL import Image
```

```python
#load the image
image = Image.open('test1.jpeg')
```

```python
#summarize some details about the image
print(image.format)
```

```python
print(image.mode)
```

## How to Convert Images to NumPy Arrays and Back

```python
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
#load image as pixel array

data = image.imread('test1.jpeg')
```

```python
#summarize shape of the pixel array
print(data.shape)
```

```python
print(data.dtype)
```

```python
#display the array of pixels as an image
plt.imshow(data)
plt.show()
```

## Load image and convert to and from NumPy array

```python
from PIL import Image
from numpy import asarray

#load the image
image = Image.open('test1.jpeg')
#convert image to numpy array
data= asarray(image)
#summarize shape
print('Shape of the data:{}'.format(data.shape))
#create Pillow image
image2 = Image.fromarray(data)
#summarize image details
print('Format:{}'.format(image2.format))
print('Mode:{}'.format(image2.mode))
print('Size:{}'.format(image2.size))
```

## Load all images in a directory

```python
import os
from matplotlib import image

#load all images in a directory
path='images'
loaded_images = []
for filename in os.listdir(path):
    # load image
    img_data = image.imread(path+"/"+filename)
    #store loaded image
    loaded_images.append(img_data)
    print('File \t\t\t Shape')
    print('{} \t\t {}'.format(filename, img_data.shape))
```

### How to Save Images to File

```python
from PIL import Image
#load the image
image = Image.open('test1.jpeg')
#save as PNG format
image.save('test1.png', format='PNG')
#load the image again and inspect the format
image2 = Image.open('test1.png')
print(image2.format)
image2.show()
```
